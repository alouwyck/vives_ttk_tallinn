In general, **array slicing** allows you to extract a portion of an array using the syntax `start:stop:step`:

- `start`: The index where the slice starts (inclusive). If omitted, it defaults to the beginning of the array.
- `stop`: The index where the slice ends (exclusive). If omitted, it defaults to the end of the array.
- `step`: The step size or stride between elements. If omitted, it defaults to 1.

Here, slice `1:5:2` will select elements starting from index 1 up to (but not including) index 5, with a step size of 2.

### 2.9. Vectorized Expressions

In numpy it is better to use vectorized expressions instead of loops!

```python
M = np.ones((1000, 1000))  # one million elements!
```

```python
from time import time

start = time()

M = M * 10  # Vectorize = multiply the numpy array by 10! You may also write M *= 10

dt_vectorized = time() - start

print(dt_vectorized, 'sec')
```

```python
start = time()

for i in range(M.shape[0]):
    for j in range(M.shape[1]):
        M[i, j] *= 10

dt_loops = time() - start

print(dt_loops, 'sec')
print()
print(f"Vectorized expression is {dt_loops / dt_vectorized :.2f} times faster!")
```

Do you want to apply logical operators to arrays with booleans, then you must use the "bitwise" operators `&`, `|` and `~`:

```python
r = np.random.rand(2, 3)
print(r)
print()

print((r > 0.25) & (r < 0.75))  # & is AND. Parentheses are compulsory!
print()
print((r <= 0.25) | (r >= 0.75))  # | is OR. Parentheses are compulsory!
print()
print(~((r <= 0.25) | (r >= 0.75)))  # ~ is NOT. Parentheses are compulsory!
```

```python
# functions any() and all()
print(np.any(r > 0.25))  # is there ANY element greater than 0.25?
print(np.all(r < 0.9))   # are ALL elements smaller than 0.9?
```

### 4.2. Training a Neural Net on the Halloween Candy Dataset

#### The Halloween Candy Dataset

```python
import pandas as pd

df = pd.read_csv('candy-data.csv')
df.head()  # show the first 5 records
```

```python
X, y = df.iloc[:, 2:].values, df.iloc[:, 1].values  # input X, output y
X[:, -1] /= 100
y = y.reshape(-1, 1)
X.shape, y.shape
```

#### Building the Neural Net

```python
num_of_nodes = [8, 5]   # number of nodes for the two hidden layer
alpha = 0.1             # learning rate
num_of_epochs = 10000   # number of epochs (= iterations)
batch_size = 17         # batch size
```

```python
g = lambda z: 1 / (1 + np.exp(-z))  # sigmoid
dgdz = lambda z: g(z) * (1 - g(z))  # sigmoid derivative
```

```python
nsamples = X.shape[0]            # number of samples
nlayers = len(num_of_nodes) + 2  # total number of layers
```

#### Initializing the Weights

```python
# initialize the weights (= model parameters)
W = []
n1 = X.shape[1] + 1
for k in num_of_nodes:
    n2 = k
    W.append(np.random.randn(n2, n1))
    n1 = k + 1
W.append(np.random.randn(y.shape[1], n1))

# verify the shapes
for i in range(len(W)):
    print(W[i].shape)
```

```python
# helper function to add row of ones on top of matrix M
def add_ones(M):
    return np.concatenate((np.ones((1, M.shape[1])), M), axis=0)
```

#### The Forward and Backward Pass

```python
# function for the forward pass
def forward(W, Xb, g):
    Z = []
    A = [add_ones(Xb.T)]
    for k in range(nlayers - 1):
        Z.append(np.dot(W[k], A[k]))
        A.append(add_ones(g(Z[k])))
    return Z, A
```

```python
# function for the backward pass
def backward(W, Z, A, yb, dgdz, alpha):
    E = A[-1][1:, :] - yb.T
    D = E * dgdz(Z[-1])
    for k in range(nlayers - 2, 0, -1):
        W[k] -= alpha * np.dot(D, A[k].T)
        D = np.dot(W[k][:, 1:].T, D) * dgdz(Z[k-1])
    W[0] -= alpha * np.dot(D, A[0].T)
    return W
```

#### Making Predictions

```python
# function to predict labels
def predict(W, X, g):
    A = add_ones(X.T)
    for k in range(nlayers - 1):
        Z = np.dot(W[k], A)
        A = add_ones(g(Z))
    return A[1:, :].T
```

```python
# function to calculate the loss function = Mean Squared Error
def MSE(W, X, g, y):
    return np.mean(np.square(predict(W, X, g) - y), axis=0)
```

#### Training the Neural Net

```python
# training the neural net!

losses = []

# indices of samples
idx = np.array(range(nsamples))

# outer loop: epochs
for k in range(num_of_epochs):

    # print
    if k % 1000 == 0:
        print(f'iteration{k}: MSE = {MSE(W, X, g, y).item()}')

    # shuffle indices
    np.random.shuffle(idx)

    # initialize mini-batch indices
    b = np.arange(batch_size)

    # inner loop: mini-batches
    for i in range(nsamples // batch_size):

        # forward pass
        Z, A = forward(W, X[idx[b], :], g)

        # backward pass
        W = backward(W, Z, A, y[idx[b], :], dgdz, alpha)

        # update mini-batch indices
        b += batch_size

    # calculate loss function
    losses.append(MSE(W, X, g, y))
```

```python
# plot loss versus epoch
plt.plot(losses);
plt.xlabel('Epoch');
plt.ylabel('MSE');
plt.grid();
```

#### Evaluating the Model

```python
# calculate accuracy
y_pred = np.round(predict(W, X, g))
print(f'Accuracy = {np.sum(y_pred == y) / nsamples:.4f}')
```
