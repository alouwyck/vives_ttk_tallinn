## What is Hydrogeology?

Hydrogeology is the science that studies **groundwater**, derived from the Greek words *hydor* (water), *ge* (earth), and *logia* (study). It focuses on how groundwater is distributed and moves through soil and rock layers in the Earth's crust.

This discipline is essential because groundwater plays a critical role in drinking water supply, agriculture, industry, and environmental sustainability. Understanding aquifers, flow dynamics, and interactions with surface water is key to managing this hidden resource responsibly.

## What is Groundwater?

Groundwater is water stored beneath the Earth's surface in pores and fractures of soil and rock. It is part of the **hydrosphere** and it forms a major component of the **hydrological cycle**, infiltrating from precipitation, storing in aquifers, and eventually discharging into rivers, lakes, and oceans. Excluding ice, groundwater is the **largest source of freshwater** on Earth, making it indispensable for ecosystems and human use.

**For humans**, groundwater provides drinking water for billions of people worldwide, irrigation water for agriculture, and industrial water for manufacturing processes. It is even used for cooling in certain industrial facilities, including some data centers that support modern technologies such as artificial intelligence, because groundwater offers a stable temperature and reliable supply.

For ecosystems, groundwater sustains **groundwater-dependent ecosystems** (GDEs), which include wetlands and springs. These ecosystems rely on subsurface water to maintain biodiversity, support plant growth, and provide habitats for wildlife. Protecting groundwater is therefore essential not only for human needs but also for ecological balance.

## What is the Hydrological Cycle?

Groundwater movement and storage are closely linked to the hydrological cycle, also called the water cycle.


The hydrological cycle describes how water continuously circulates through the Earth's atmosphere, surface, and subsurface. It is a closed system powered mainly by solar energy and gravity. This cycle ensures that water is constantly recycled, making life on Earth possible.

Water begins its journey when heat from the sun causes evaporation from oceans, lakes, and rivers. Plants also contribute through transpiration, releasing water vapor into the atmosphere. This vapor rises, cools, and condenses into clouds. When the clouds become heavy enough, precipitation occurs in the form of rain, snow, or hail.

Once water reaches the ground, several processes take place. Some of it runs off across the surface into rivers and streams, eventually returning to the oceans. Another portion infiltrates the soil, replenishing groundwater stored in aquifers. This infiltration is crucial because groundwater acts as a long-term reservoir, feeding springs and maintaining river flow during dry periods. Water can also freeze and become part of glaciers, storing it for centuries.

The water cycle is dynamic and interconnected. Changes in one part, such as reduced infiltration due to urbanization, can affect groundwater recharge and river flow. Climate change also influences the cycle by altering evaporation rates and precipitation patterns.

**Key stages of the water cycle:**

*   **Evaporation and transpiration:** Water moves from surface and plants to the atmosphere. When these two processes are considered together, the term **evapotranspiration** is used.
*   **Condensation:** Water vapor forms clouds.
*   **Precipitation:** Rain or snow returns water to the surface.
*   **Runoff:** Water flows over land into rivers and oceans.
*   **Infiltration and percolation:** Water enters the soil (infiltration) and recharges aquifers (percolation).
*   **Storage:** Water is stored in oceans, lakes, glaciers, and groundwater.

Understanding this cycle is essential for hydrogeology because groundwater is not isolatedit is part of this global system. Managing groundwater sustainably requires considering how recharge and discharge interact with surface water and climate.

## What is an Aquifer?

An aquifer is a geological formation capable of storing and transmitting groundwater. **Unconfined or phreatic** aquifers have a water table exposed to atmospheric pressure, while **confined** aquifers are bounded by less permeable layers, keeping water under pressure.

Other formations include **aquitards**, which slow groundwater flow, and **aquicludes**, which are nearly impermeable. Materials like sand and gravel typically form good aquifers, whereas clay often acts as an aquitard. Fractured rocks can often act as aquifers, while solid, unfractured rocks usually behave as aquicludes.


## What is a Groundwater Reservoir?

In most cases, a groundwater reservoir is not just a single aquifer. It often consists of a **sequence of aquifers and aquitards** stacked on top of each other, while the bottom is typically bounded by an impermeable layer called an aquiclude. This layered structure controls how water moves and is stored in the underground.

In hydrogeology, we treat a groundwater reservoir as an **open system**, applying **systems thinking** to understand its behavior. This means we look at how the reservoir interacts with surface water, climate, land use, and human activities. **Measurements** reveal **events** such as droughts or contamination incidents, which can have significant impacts on water availability and quality.

Modern tools allow us to go further. Using machine learning, we can detect **patterns** in large datasets of groundwater measurements and make predictions about future trends and **behavior** of the groundwater system. These are **statistical models**  often called black-box models  because they can predict outcomes without explaining the underlying processes.

However, if we truly want to understand the system, we need **mathematical models** that **simulate** the physical and chemical **processes** governing groundwater flow and quality. These white-box models are process-based and they are essential for improving our scientific knowledge and for supporting sustainable **groundwater management and policy** decisions. They provide the foundation for evidence-based strategies that balance human needs with environmental protection.

## What is the Hydraulic Head?

We measure groundwater levels in wells, and this measurement is called the **hydraulic head**. Hydraulic head represents the total mechanical energy of groundwater at a specific point, expressed as an equivalent height of water. It determines the direction of groundwater flow, which always moves from areas of high head to areas of low head.

Hydraulic head consists of two components:

- **Elevation head**: The height of the measurement point above a reference level (usually sea level).
- **Pressure head**: The height of a water column that corresponds to the pressure exerted by groundwater at that point.

The formula for hydraulic head is:

$$
h = z + \frac{p}{\gamma}
$$

Where:
- $h$ = hydraulic head [L]
- $z$ = elevation head [L]
- $p$ = pressure [F/L^2]
- $\gamma$ = specific weight of water [F/L^3], often expressed as $\rho g$ (density  gravitational acceleration)

In practice, hydraulic head is measured using **observation wells** or **piezometers**, referencing either ground level or sea level. These measurements are essential for mapping water tables, determining flow directions, and designing wells.

## What is Darcy' Law?

**Groundwater flow** is **laminar** and it follows Darcy's Law, which states that the **flow rate** through a **porous** medium is proportional to the **hydraulic gradient** and the material's **hydraulic conductivity**. This principle forms the foundation of hydrogeology. Hydraulic conductivity measures how easily water moves through soil or rock, while the hydraulic gradient represents the slope of the water table in a phreatic aquifer or the potentiometric surface in a confined aquifer.

The one-dimensional, differential form of Darcy's Law is:

$$
q = -K \dfrac{dh}{dx}
$$

Where:
- $q$ = **specific discharge** or flow rate [L/T]
- $K$ = **hydraulic conductivity** [L/T]
- $\dfrac{dh}{dx}$ = **hydraulic gradient** [-]

For a finite-difference approximation over a distance $\Delta x$, Darcy's Law becomes:

$$
q \approx -K \dfrac{\Delta h}{\Delta x}
$$

Where:
- $\Delta h$ = difference in hydraulic head between two points [L]
- $\Delta x$ = distance between these two points [L]

### Example

Consider two observation wells, A and B, located 100 meters apart. The groundwater level (hydraulic head $h$) in well A is measured at 50 meters, while in well B it is 40 meters. The aquifer is sandy, with a known hydraulic conductivity $K$ of 10 meters per day. To calculate the specific discharge $q$ using Darcy's Law, we determine the hydraulic gradient from the difference in head and the distance between the wells, and we apply the above formula:


```python
K = 10   # aquifer conductivity (m/d)
hA = 50  # hydraulic head (m) in well A
hB = 40  # hydraulic head (m) in well B
L = 100  # distance (m) between well A and B

q = -K * (hB - hA) / L  # Darcy's law

print(f'The flow rate is {q:.2f} m/d')
```

## What is the Continuity Equation?

The continuity equation expresses the principle of **conservation of mass** in groundwater systems. For groundwater, which is nearly **incompressible**, this means working with **volumes**.

If the system is in a **steady state**, groundwater levels do not change over time, meaning that inflow (or recharge) equals outflow (or discharge). In **transient conditions**, groundwater levels change significantly over time, and therefore, changes in storage must also be considered.

**Sources** of inflow include injection wells and recharge from precipitation, irrigation, and infiltrating rivers and lakes. Outflows occur through pumping wells, subsurface drainage, discharging rivers, and evapotranspiration. These outflows are often referred to as **sinks**.

For groundwater, the continuity equation can be written as:

$$
\text{inflow} + \text{storage change} = \text{outflow}
$$

Rearranging gives:

$$
\text{outflow} - \text{inflow} = \text{storage change}
$$

The unit of $\text{outflow}$, $\text{inflow}$, and $\text{storage change}$ is volume per unit of time. So, we are describing a **rate**, because it expresses how much volume flows or changes per unit of time. As groundwater flow is relatively slow, it is common to use cubic meters per day (m/day).

As we will discuss later, this relationship is fundamental for groundwater modeling. In steady-state conditions, the storage change term becomes zero, so inflow equals outflow. In transient models, the storage term accounts for changes in hydraulic head over time.

### Example

Consider a confined aquifer with a pumping well extracting groundwater at a rate of 100 m/day. The aquifer is bounded below by an impermeable layer and above by an aquitard.

At the start of pumping, all the water comes from the aquifer's storage because there is no immediate inflow from the aquitard. In this initial phase, the storage change equals the pumping rate (100 m/day), while inflow is essentially zero.

Over time, as pumping continues, leakage from the overlying aquitard begins to contribute water to the well. Eventually, a state of *dynamic equilibrium* is reached: the inflow from the aquitard equals the pumping rate (100 m/day), and the change in storage becomes negligible. At this point, the system behaves as if it were in steady state.


## What is a Pumping Well?


A pumping well is a **drilled hole equipped with a pump** to extract groundwater. Pumping lowers the groundwater level and creates a **cone of depression** around the well, altering the natural flow system. Such wells are widely used for water supply and irrigation.

A typical well construction includes:

- A **casing**, which is the pipe that lines the borehole to prevent collapse and protect the water from contamination.
- A **screen**, which is a perforated section of the casing that allows water to enter the well while keeping out sediment.
- Often a **gravel pack**, placed around the screen to improve filtration.

These components ensure efficient **groundwater extraction** and long-term well stability.

## What is Groundwater Modeling?

Groundwater modeling simulates groundwater **flow** and groundwater **quality** using **computer models**. These models help predict how aquifers respond to changes such as pumping or climate variations and they are used for water management, contamination studies, and environmental impact assessments. They range from simple to complex and can be one-, two-, or three-dimensional.

Modeling relies on real-world data and assumptions about subsurface conditions. The **governing equations** for groundwater flow combine **Darcy's Law** and the **continuity equation**, resulting in the groundwater flow equation, which accounts for hydraulic conductivity, storage change, and changes in hydraulic head over time.

In its simplest one-dimensional form, the **groundwater flow equation** for a homogeneous aquifer is:

$$
T \dfrac{\partial^2 h}{\partial x^2} = S \dfrac{\partial h}{\partial t}
$$

Where:
- $h$ = hydraulic head [L]
- $T$ = aquifer **transmissivity** [L/T], defined as $T = K \cdot D$, or hydraulic conductivity times aquifer thickness
- $S$ = aquifer **storativity** [-], a dimensionless property that represents the volume of water released from storage per unit surface area of aquifer per unit decline in hydraulic head
- $x$ = spatial coordinate [L]
- $t$ = time [T]

This **partial differential equation** expresses that changes in hydraulic head over time are driven by spatial variations in head and the properties of the aquifer. It applies the **continuity equation**: the left-hand side represents the difference between outflow and inflow using **Darcy's Law**, while the right-hand side represents the change in storage using storativity. In case of steady-state flow, the right-hand side is zero.

Sources and sinks, such as wells and rivers, are incorporated into the model as additional mathematical equations called **boundary conditions**. These conditions ensure that the governing flow equation has a unique solution. When the problem is defined in this way, it becomes a **boundary value problem**. Solving a boundary value problem can be done either analytically, using exact mathematical solutions, or numerically, by approximating the solution with computational methods.

**Analytical solutions** provide exact formulas, such as the Thiem-Dupuit equations for steady flow to wells, while **numerical solutions** approximate reality by discretizing the domain and using iterative methods like finite differences or finite elements. In general, numerical models are more advanced than analytical models.

Industry-standard tools include [MODFLOW](https://www.usgs.gov/mission-areas/water-resources/science/modflow-and-related-programs), which is a numerical model, and its Python interface [FloPy](https://github.com/modflowpy/flopy). In this notebook, we will use [TimML](https://timml.readthedocs.io/en/latest/index.html), a Python package to build relatively simple analytical models. Code examples will be given and discussed later on.

## What is an Axisymmetric Model?

Axisymmetric models are a special type of groundwater flow model commonly used to simulate **flow toward a pumping well**. Instead of using Cartesian coordinates $(x, y, z)$, these models employ **cylindrical coordinates** $(r, \theta, z)$. By assuming radial symmetry, the problem is simplified by omitting the -dimension, since the flow pattern is identical in every direction around the well.

**Radial symmetry** means that groundwater flows toward the well in a perfectly circular pattern, without any directional preference. This assumption is valid under **specific conditions**:

- The pumping well is perfectly vertical and cylindrical.
- The aquifer is homogeneous and horizontally extensive.
- The flow toward the well is purely radial, without significant vertical or lateral variations.

These models are historically important because they were among the first groundwater flow models ever developed. **Analytical solutions** for axisymmetric flow, such as the Thiem-Dupuit formulas, were introduced in the second half of the 19th century and are still widely used.

Before the advent of computers, many analytical solutions were derived for various axisymmetric configurations. **Numerical models** only began to dominate groundwater modeling in the 1960s with the rise of digital computing.

Axisymmetric models remain useful today for conceptual understanding and for quick analytical calculations in **well hydraulics**.

## The Thiem-Dupuit Formulas

The French engineer **Jules Dupuit** derived these formulas as early as 1857, just one year after Henri Darcy formulated his famous law. Independently of the French work, the German engineer **Adolf Thiem** developed the same formula for confined flow, which was later applied in practice by his son **Gnther Thiem** for the interpretation of **pumping tests**. These are field experiments where a well is pumped at a constant discharge rate and the drawdown (lowering of hydraulic head) is measured in observation wells at various distances to determine aquifer properties, in this case the hydraulic conductivity of the aquifer.


The Thiem-Dupuit formulas simulate **steady-state radial one-dimensional flow toward a pumping well in a homogeneous aquifer**.

There are two variants:

- one for **constant aquifer thickness** (confined aquifer  linear problem),
- one for **variable thickness** (unconfined aquifer  nonlinear problem).

The **pumping rate** is constant, and the **radius of influence**, this is the distance at which the drawdown is practically zero, is assumed to be known in advance. These formulas are commonly used to analyze pumping test data, often through semi-logarithmic plots of drawdown versus radial distance.

This is the formula for **confined flow**, commonly referred to as the **Thiem equation**:

$$
s(r) = \frac{Q}{2 \pi K D} \ln \left( \frac{R}{r} \right)
$$

Where:

- $s(r)$ = drawdown [L] at distance $r$  
- $r$ = radial distance [L] from pumping well  
- $Q$ = pumping rate [L/T]
- $K$ = hydraulic conductivity [L/T]  
- $D$ = constant saturated thickness [L]  
- $R$ = radius of influence [L]  


This is the formula for **unconfined flow**, known as the **Dupuit formula**:

$$
s(r) = h_0 - \sqrt{ h_0^2 - \frac{Q}{\pi K} \ln \left( \frac{R}{r} \right) }
$$

Where:
- $s(r)$ = drawdown [L] at distance $r$  
- $r$ = radial distance [L] from pumping well  
- $h_0$ = initial water table elevation [L]  
- $Q$ = pumping rate [L/T]
- $K$ = hydraulic conductivity [L/T]  
- $R$ = radius of influence [L]


For confined aquifers, the aquifer thickness $D$ remains constant, making the problem linear, whereas in unconfined aquifers the saturated thickness depends on the hydraulic head $h$, resulting in a nonlinear problem. Consequently, the first formula does not depend on the initial head $h_0$, while the second formula does.

### Example

Consider a pumping well extracting groundwater at constant rate $Q$ equal to 1000 m/d from a homogeneous aquifer with thickness $D$ equal to 20 m and hydraulic conductivity $K$ equal to 10 m/d. The aquifer thickness is determined by the initial water table elevation $h_0$. The radius of influence $R$ of the pumping well equals 500 m.

Let's assign the given numbers to Python variables with the same name:

```python
Q  = 1000    # pumping rate (m/d): Q > 0 -> extraction
R  = 500     # radius of influence (m)
K  = 10      # conductivity (m/d)
h0 = D = 20  # initial head = initial aquifer thickness (m)
T  = K * D   # aquifer transmissivity (m/d)
```

Note that the aquifer transmissivity $T$ equals $KD$.

If we want to know the drawdown $s$ at a distance $r$ of 1 m from the pumping well, we simply apply the two formulas:

```python
# calculate the drawdown at a distance of 1 m from the well
r = 1

# drawdown s according to the Thiem formula:
s1 = Q / 2 / np.pi / T * np.log(R / r)

# drawdown s according to the Dupuit formula:
s2 = h0 - np.sqrt(h0**2 - Q / np.pi / K * np.log(R / r))

# print the result
print(f'Drawdown according to Thiem:  s = {s1:.2f} m')
print(f'Drawdown according to Dupuit: s = {s2:.2f} m')
```

The Dupuit formula results in a larger drawdown $s$ because it assumes that the aquifer thickness $D$ equals the hydraulic head $h$. This reduces the effective thickness and thus the aquifer transmissivity, which is compensated by a greater drawdown and increased flow toward the well to maintain the constant pumping rate.

If we want to repeat these calculations for different distances $r$, then it is a good programming practice to define a **Python function** for each formula:

```python
def thiem(r, T, Q, R):
    """
    Calculate drawdown at given distances r according to the Thiem formula for steady confined flow.

    Parameters
    ----------
    r : array_like
      Radial distances [L], which should be smaller than `R`.
    T : float
      Aquifer transmissivity [L/T].
    Q : float
      Pumping rate [L/T] of the well.
    R : float
      Radial distance [L] of the outer aquifer boundary.

    Returns
    -------
    s : ndarray
      Drawdown [L] at given distances `r`. The shape of `s` is the same as the shape of `r`.
    """
    r = np.array(r)
    return Q / 2 / np.pi / T * np.log(R / r)
```

```python
def dupuit(r, K, h0, Q, R):
    """
    Calculate drawdown at given distances r according to the Dupuit formula for steady unconfined flow.

    Parameters
    ----------
    r : array_like
      Radial distances [L], which should be smaller than `R`.
    K : float
      Aquifer conductivity [L/T].
    h0 : float
       Initial hydraulic head [L] which is the initial aquifer thickness before pumping.
       `h0` is also the constant head at the outer aquifer boundary at distance `R`.
    Q : float
      Pumping rate [L/T] of the well (which is positive in case of extraction).
    R : float
      Radial distance [L] of the outer aquifer boundary.

    Returns
    -------
    s : ndarray
      Drawdown [L] at given distances `r`. The shape of `s` is the same as the shape of `r`.
    """
    return h0 - np.sqrt(h0**2 - Q / np.pi / K * np.log(R / r))
```

Now it is straightforward to calculate the drawdown $s$ for a range of distances $r$ logarithmically spaced between 0.01 m and 500 m (which is the radius of influence $R$):

```python
r = np.logspace(np.log10(0.01), np.log10(R), 100)  # log-spaced radial distances (m)
s1 = dupuit(r=r, Q=Q, K=K, R=R, h0=h0)  # drawdown (m) according to the Dupuit formula
s2 = thiem(r=r, Q=Q, T=T, R=R)          # drawdown (m) according to the Thiem formula
```

Note that we make use of **NumPy arrays**, which are efficient containers for numerical data in Python. They allow fast mathematical operations on entire datasets without explicit loops, making calculations such as drawdown or transmissivity much more convenient and performant.

Let's plot the corresponding semi-logarithmic **distance-drawdown graph**:

```python
plt.semilogx(r, s1, 'b-', label='unconfined (Dupuit)');  # drawdown s (m) according to the Dupuit formula
plt.semilogx(r, s2, 'b:', label='confined (Thiem)');     # drawdown s (m) according to the Thiem formula
plt.xlabel('r (m)');
plt.ylabel('s (m)');
plt.legend();
plt.grid();
```

We clearly see that the Dupuit formula results in higher calculated drawdown near the well.

Although previous semi-logarithmic graph is more practical, for instance, to analyze pumping test data, it's also possible to visualize the actual **cone of depression**:

```python
plt.figure(figsize=(10, 2));
plt.plot(-r, -s1, 'b-');
plt.plot(r, -s1, 'b-');
plt.plot(-r, -s2, 'b:');
plt.plot(r, -s2, 'b:');
plt.grid();
```

As you can see, using semilog plots is more convenient because they zoom in on the drawdowns near the pumping well. Close to the pumping well, drawdowns are the largest and the hydraulic gradient is steepest. Farther away, the drawdowns become much smaller and approach zero near the radius of influence.

## TimML

It is also possible to use TimML to simulate **steady flow to a pumping well**. [TimML](https://timml.readthedocs.io/en/latest/) is a Python package for the modeling of steady-state multi-layer groundwater flow with **analytic elements** developed by [Mark Bakker](https://www.tudelft.nl/citg/over-faculteit/afdelingen/watermanagement/medewerker/staff-hydrology/academic-staff/profdrir-mark-bakker), Professor in Groundwater Engineering at the Technical University of Delft in the Netherlands.

First, we need to install TimML:

```python
!pip install timml -q
```

After installing the TimML package, we import it:

```python
import timml
```

The first step in a TimML model is to create a **model** and define the **aquifer properties**. Here we use the `ModelMaq` class and specify the hydraulic conductivity `kaq`, and the top and bottom of the aquifer, which are assigned to `z`. The `ModelMaq` class returns a `model` object:

```python
model = timml.ModelMaq(
    kaq=K,     # hydraulic conductivity of the aquifer (m/d)
    z=[D, 0],  # top and bottom of the aquifer (m)
)
```

Once the model is created, we can add analytic elements to the model, such as a **pumping well**. Here, we need a `Well` with pumping rate equal to `Q`. To keep it simple, we define the well at (0,0), and set the well radius `rw` to the first radial distance in array `r`:

```python
well = timml.Well(
    model=model,  # model to which element is added
    xw=0,         # x-location of well (m)
    yw=0,         # y-location of well (m)
    Qw=Q,         # discharge of well (m/d), positive for extraction
    rw=r[0],      # well radius (m), we take the first element of r
)
```

 Recall that we also have to define a `Constant` head at distance `R`, the **radius of influence**:

```python
constant = timml.Constant(
    model=model,  # model to which element is added
    xr=R,         # x-location of fixed head (m)
    yr=0,         # y-location of fixed head (m)
    hr=h0,        # fixed head (m)
)
```

Now we're ready to **solve the model**, which means that any unknown parameters in the model are computed.

```python
model.solve()
```

Calling method `model.head()` gives the head at a specified point (x,y). **Drawdown** at distances `r` is calculated as the difference between the initial head `h0` and the head $h$ during pumping, or:

> $s = h_0 - h$

```python
h = np.array([model.head(x=r, y=0) for r in r])
s3 = h0 - h
```

This gives virtually the same result as the **Thiem solution** we've calculated earlier:

```python
plt.semilogx(r, s2, 'k-', label='Thiem');
plt.semilogx(r, s3, 'r:', label='TimML');
plt.legend();
plt.xlabel('r (m)');
plt.ylabel('s (m)');
plt.grid();
```

Another nice feature of TimML is method `contour` to create a **contour plot** of the hydraulic head $h$. A contour plot displays lines of equal hydraulic head, and in this case, these lines are perfect circles because the problem is **axially symmetric**.

```python
model.plots.contour(
    win=[-R/2, R/2, -R/2, R/2],  # window to contour [xmin, xmax, ymin, ymax]
    ngr=100,                     # number of points where to compute the head
    levels=np.arange(20),        # levels to be contoured
    labels=True,                 # add labels along the contours
    decimals=1,                  # print labels with 1 decimal place
    legend=True,                 # add a legend
    figsize=(5, 5),              # specify a figure size of 5 by 5 inches
);
```

Is it possible to simulate a Dupuit model using TimML? The answer is no, because TimML only supports confined flow, or more accurately, horizontal layers with constant thickness and constant hydraulic conductivity.

The various steps you need to take with TimML may seem somewhat cumbersome at first glance, but this is because TimML is a generic model for simulating groundwater flow in groundwater systems consisting of multiple aquifers and aquitards. In addition to wells and constant heads, more complex elements can also be added to a model, such as line sinks and areal sinks, which can be used to simulate interactions with rivers and infiltration areas.

For instance, we assumed there is no groundwater flow in the aquifer before the pumping starts, which is very unlikely. TimML provides a `UFlow` element to add uniform flow to the aquifer:  

```python
flow = timml.Uflow(
    model=model,  # model to which element is added
    slope=0.01,   # hydraulic gradient = head difference / distance (in direction of flow)
    angle=0,      # direction of uniform flow, straight East is zero degrees
)
```

**Uniform flow** refers to a situation where groundwater moves through the aquifer at a constant velocity and in a single direction, typically caused by a regional hydraulic gradient. In other words, the head decreases linearly across the aquifer, and flow lines are parallel.

We can calculate the flow rate $q$ of this uniform flow using Darcy's law. The `slope` input parameter of the `Uflow` element actually defines the **hydraulic gradient** $\Delta h / \Delta x$, which is set to `0.01` in this case.

Let's solve the model again so it accounts for the added uniform flow:

```python
model.solve()
```

The contour plot looks different now:

```python
model.plots.contour(
    win=[-R, R, -R, R],        # window to contour [xmin, xmax, ymin, ymax]
    ngr=100,                   # number of points where to compute the head
    levels=np.arange(20, 31),  # levels to be contoured
    labels=True,               # add labels along the contours
    decimals=1,                # print labels with 1 decimal place
    legend=False,              # don't add a legend
    figsize=(5, 5),            # specify a figure size of 5 by 5 inches
);
```

Remember that we defined a constant head of 20m at a point 500m east of the pumping well along the x-axis. As expected, the hydraulic head there is indeed 20m. There is a uniform flow increasing from east to west, so at 500m west of the well the head is 30m. This is easy to explain because the hydraulic gradient is 0.01, meaning the head rises by 1cm per meter, or 10m per kilometer. Uniform flow normally produces parallel streamlines  in this case running north to south  but the pumping well disturbs this pattern. At the well location (0,0), we clearly see a local cone of depression.

As already mentioned, it is also possible to include the effect of a river or a canal in the model by using, for example, the `HeadLineSink` element. Here we add an **east-west oriented canal** located 100m north of the pumping well. The `HeadLineSink` element represents the canal as a straight line between two points, `(x1, y1)` and `(x2, y2)`. The canal has a fixed water level `hls`, which we will set to 20m. We can also specify a width `wh`; here we take 10m. Finally, we define a hydraulic resistance `res`. Simply put, this resistance controls the exchange of water between the canal and the aquifer, and the higher the resistance, the more difficult it is for water to move between them.

```python
river = timml.linesink.HeadLineSink(
    model,            # model to which element is added
    x1=-500, y1=100,  # coordinates of the first point of the canal
    x2=500, y2=100,   # coordinates of the second point of the canal
    hls=20.0,         # constant water level (m) in the canal
    wh=10,            # width (m) of the canal
    res=10,           # hydraulic resistance (d) between canal and aquifer
)
```

Solving the model to take into account the newly added element:

```python
model.solve()
```

The contour plot clearly shows the draining effect of the canal:

```python
model.plots.contour(
    win=[-R, R, -R, R],        # window to contour [xmin, xmax, ymin, ymax]
    ngr=100,                   # number of points where to compute the head
    levels=np.arange(20, 31),  # levels to be contoured
    labels=True,               # add labels along the contours
    decimals=1,                # print labels with 1 decimal place
    legend=False,              # don't add a legend
    figsize=(5, 5),            # specify a figure size of 5 by 5 inches
);
```

Recall that the contour plot shows lines of equal hydraulic head. To visualize groundwater movement within the aquifer, we add so-called **pathlines** to the plot. These are shown in orange and are perpendicular to the contour lines, indicating the direction of groundwater flow.

```python
# contour plot
model.plots.contour(
    win=[-R, R, -R, R],        # window to contour [xmin, xmax, ymin, ymax]
    ngr=100,                   # number of points where to compute the head
    levels=np.arange(20, 31),  # levels to be contoured
    labels=True,               # add labels along the contours
    decimals=1,                # print labels with 1 decimal place
    legend=False,              # don't add a legend
    figsize=(5, 5),            # specify a figure size of 5 by 5 inches
)

# pathlines
xstart = np.hstack((
    np.arange(-400, 501, 100),
    -500 * np.ones(21),
    np.arange(-400, 501, 100),
))
ystart = np.hstack((
    500 * np.ones(10),
    np.arange(-500, 501, 50),
    -500 * np.ones(10),
))
model.plots.tracelines(
    xstart=xstart,                 # x-locations of starting points
    ystart=ystart,                 # y-locations of starting points
    zstart=np.zeros(len(xstart)),  # z-locations of starting points
    hstepmax=20,                   # maximum horizontal step size
    color="C1",                    # color of lines
);
```

Remember that groundwater flows from areas of high hydraulic head to areas of low head. Here, this is clearly from west to east because of the uniform flow element we've added to the model. Moreover, the pathlines clearly reveal that groundwater also flows toward the canal, meaning the canal drains the aquifer. And of course, water flows toward the pumping well as well.

In this notebook, we introduced the fundamental concepts and principles of hydrogeology: **groundwater**, **the hydrological cycle**, **groundwater reservoirs**, **aquifers and aquitards**, **confined versus unconfined**, **hydraulic head**, **Darcy's law**, **pumping wells**, and the basics of **groundwater modeling**.

We then demonstrated the **Thiem-Dupuit equations**, which provide analytical solutions for steady-state radial flow toward a pumping well in confined and unconfined aquifers.

Finally, we built a simple analytical model using **TimML**, simulating a **pumping well** in an aquifer with **uniform regional flow** and a nearby **canal**. TimML allows the creation of more complex models by adding additional elements such as line sinks, recharge zones, or multiple wells.
